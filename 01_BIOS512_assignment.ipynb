Using the `penguins` dataset, make a scatter chart with `bill_length_mm` on the x-axis and `bill_depth_mm` on the y-axis.

```r
options(repr.plot.width=5, repr.plot.height=4)
pen_chart= ggplot (penguins, aes (x=bill_length_mm, y= bill_depth_mm)) + geom_point(na.rm=TRUE)
pen_chart
```

Make the same plot but *encode* the `species` column with color.

```r
pen_chart2= ggplot (penguins, aes (x=bill_length_mm, y= bill_depth_mm, color=species)) + geom_point(na.rm=TRUE)
pen_chart2
```
